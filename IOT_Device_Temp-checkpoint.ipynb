##### Technical Details:

noted_date : date and time of reading

temp : temperature readings

out/in : whether reading was taken from device installed inside or outside of room?

## Importing Required Libraries and files

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

file_loc = "C:/Users/p/IOT Devices/IOT-Temp.csv"
df = pd.read_csv(file_loc)
df.head()
```

### We can see that id and room_id/id columns will not prodcue any valuable insights and there is repetition of data in rows


```python
df.info()
```

## Cleaning Data

#### It can be seen that there are no null value and dtype of noted_date is object which is incorrect
Renaming out/in column to out_in for easy use with attribute referencing

Correcting the dtype

Dropping redundant columns and duplicate rows

```python
df = df.rename(columns={'out/in':'out_in'})
df['date_time'] = pd.to_datetime(df.noted_date)

cols = ['id', 'room_id/id', 'noted_date']
df = df.drop(columns=cols, axis=1)

rows = ['temp','date_time','out_in']
df.drop_duplicates(subset=rows, keep= False, inplace=True)
```

#### We can see that the number of rows has reduced considerably. Only 24% of the rows are non-repetetive

```python
df.describe()
```

## Visualizing the variation of device temperature with time using scatter plots as there are multiple readings collected in every day

```python
y = df.temp
x1 = df.date_time.dt.hour
x2 = df.date_time.dt.day
x3 = df.date_time.dt.week
x4 = df.date_time.dt.month

fig = plt.figure(figsize=(15,15))
plt.subplot(221)
ax1 = sns.scatterplot(x=x1,y=y,hue='out_in',data=df)
ax1.set(xlabel='Hour', ylabel='Temp(Celsius)', title='Hour wise Plot')
plt.subplot(222)
ax2 = sns.scatterplot(x=x2,y=y,hue='out_in',data=df)
ax2.set(xlabel='Day', ylabel='Temp(Celsius)', title='Day wise Plot')
plt.subplot(223)
ax3 = sns.scatterplot(x=x3,y=y,hue='out_in',data=df)
ax3.set(xlabel='Week', ylabel='Temp(Celsius)', title='Week wise Plot')
plt.subplot(224)
ax4 = sns.scatterplot(x=x4,y=y,hue='out_in',data=df)
ax4.set(xlabel='Month', ylabel='Temp(Celsius)', title='Month wise Plot')

```

### Observations and Inference :

*Orange dots represents device outside the room

*Blue dots represents device inside the room

It is clear from the scatter plots that there is high variance in temperature of external device as compared to the temperature of the device kept inside a room.

We can see that the hottest month during the year is September and there is a spike in temperature for internal device during this period mid week in the afternoon.(Outliers in the plot)

The device placed outside the room also records the highest temperature during this period.

##### External Device:

The highest device temperature noted is 51 (deg celsiuis).

The coolest device temperature noted is between 25 to 30 (deg celsiuis).

##### Internal Device:

The highest device temperature noted is around 40 (deg celsiuis).

The coolest device temperature noted is 22 (deg celsiuis).

From these observations it can be inferred that the devices used in external environments need more effective and faster cooling methods to ensure the device is at lower eisk of failure.

Keeping in mind the growing challenge with global warming, it is necessary to manufacture devices which produce less heat compared to the existing ones.
